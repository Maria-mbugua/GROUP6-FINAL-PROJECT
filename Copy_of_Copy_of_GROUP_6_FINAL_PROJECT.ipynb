# MERGING DATASETS

```python
# Grouping public schools per city and obtaining total enrollment per city and total number of schools in each city

public_schools.CITY = public_schools.CITY.str.strip()
public_schools.COUNTRY = public_schools.COUNTRY.str.strip()
```

```python
# total enrollment and number of schools per city

city_enroll_schools = public_schools.groupby('CITY').agg(ENROLLMENT = ('ENROLLMENT', 'sum'), SCHOOLS = ('NAME', 'count'))
city_enroll_schools
```

```python
# Removing cities with enrollment = 0

city_enroll_schools = city_enroll_schools[city_enroll_schools['ENROLLMENT']>0]

city_enroll_schools.head()
```

```python
# merging the dataset for population and new dataset for enrollment per city

merge_1 = pd.merge(city_enroll_schools, city_pop, left_index = True, right_on = 'CITY')
merge_1
```

```python
merge_1.to_csv
```

```python
# Group dataset with enrollment and population state

state_enroll_schools = merge_1.groupby('STATE').agg(ENROLLMENT = ('ENROLLMENT', 'sum'), SCHOOLS = ('SCHOOLS', 'sum'), POPULATION = ('POPULATION','sum'))
state_enroll_schools.head()
```

```python
# Merging new dataset with schools and enrollments per state, with teachers by state dataset

merge_2 = pd.merge(state_enroll_schools, state_teach, left_index = True, right_on = 'STATE')
merge_2.head()
```

```python
merge_2.drop(merge_2[merge_2.TEACHERS == '‡'].index, inplace=True)

merge_2.head()
```

```python
merge_2['TEACHERS'] = merge_2['TEACHERS'].astype(int) 
merge_2.dtypes
```

```python
# objective 1: To identify the states with the least number of teachers.

merge_2['TEACHERS'].min()
```

```python
# objective 2: To Identify the city with the highest ratio of enrollments per school 
# (the higher the ratio, the more schools needed in that city) 

city_enroll_schools['RATIO'] = city_enroll_schools['SCHOOLS']/city_enroll_schools['ENROLLMENT']
city_enroll_schools.sort_values(by = 'RATIO', ascending = False).head(20)
```

```python
# objective 3: To analyze the student teacher ratio per state.

merge_2['RATIO2'] = merge_2['TEACHERS']/merge_2['ENROLLMENT']
merge_2.sort_values(by = 'RATIO2', ascending = True).head()

# The States with the lowest ratio teachers/student are the ones that should be focused on to hire more teachers 
```

```python
# objective 4: To identify the number of schools in each level per city

public_schools.groupby(['STATE_ABR','LEVEL']).agg(Number_of_SChools = ('NAME', 'count')).head(20)
```
